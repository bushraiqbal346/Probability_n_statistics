```python
##plotting weights
#create a figure
plt.figure(figsize=(20,5))
#add a line renderer with legendand line and thickness
x = range(data['RIDRETH1'].size)
y = data['RIDRETH1']
plt.scatter(x, y , marker ='x' )

#also need to make figue readable --very important from stats prespective
plt.title('Weight of indiviauls')
plt.xlabel("ID of indiviuals")
plt.ylabel("Weights (kg)")
plt.grid()
plt.show()
```
